```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
data = pd.read_csv('input.csv', header=0)

data["Labeling"] = LabelEncoder().fit_transform(data["Labeling"])

print(data["Labeling"])

```

```python
X = data.drop(["Labeling"], axis=1).to_numpy()
o = np.ones((X.shape[0], 1))
np.concatenate([o, X], axis=1)
```

```python
Y = data["Labeling"].to_numpy()
```

```python
def sigmoid(z):
    return 1. / (1+np.exp(-z))

def loss_function(y_hat, y):
    return (-y * np.log(y_hat) - (1 - y) * np.log(1 - y_hat))

def gradient(x, y_hat, y):
    return np.dot(x.T, (y_hat - y))

def predict(x, theta):    
    z = np.dot(x, theta)
    y_hat = sigmoid(z)
    return y_hat

def gradient_descent(x, y, theta, lr, epochs):
    for epoch in range(epochs):
        for i in range(x.shape[0]):
            xi=x[i]
            yi=y[i]
            y_hat=predict(xi, theta)
            loss = loss_function(y_hat, yi)
            grad = gradient(xi, y_hat, yi)
            theta = theta - lr * grad
    return theta       
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)

theta =  np.random.uniform(size=X_train.shape[1])
theta = gradient_descent(X_train, Y_train, theta, 0.01, 1000)
```
